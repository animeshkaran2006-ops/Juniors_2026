```python
#Q2
num1 = int(input("Enter the first number"))
num2 = int(input("Enter the second number"))

for i in range(num1,num2+1):
    prime = True
    for j in range(2,int(i**1/2)+1):
        if i%j == 0:
            prime = False
    if prime:
        print(i)
```

```python
###### Q3
A =  np.random.choice(range(1, 11), size=(3,3), replace=False)
B =  np.random.choice(range(1, 11), size=(3,3), replace=False)
print("A\n", A,'\n--------')
print("B\n",B,'\n--------')

#Addition of matrices
C = A+B
print("#Addition of matrices\n",C,'\n--------')

#Element-Wise Multiplication
D = A*B
print("#Element-Wise Multiplication\n",D,'\n--------')

#Dot Product
print("#Dot Product\n",np.dot(A,B),'\n--------')

#Determinant
print("Determinant of A is", np.linalg.det(A))
```

```python
#Q7
import datetime
import matplotlib.dates as mdates

base = datetime.date(2025, 1, 1)
time_labels = [base + datetime.timedelta(days=x) for x in range(31)]

t = np.arange(len(time_labels))
temperature = 15 + 10 * np.sin(2 * np.pi * t / 31)

ts = pd.Series(data=temperature, index=time_labels, name='Temperature (°C)')

fig, ax = plt.subplots()
ax.plot(ts.index, ts.values,label='Daily Temperature')

ax.set_title('Daily Temperatures January 2025')
ax.set_xlabel('Date')
ax.set_ylabel('Temperature (°C)')
ax.grid()
plt.xticks(rotation=45)

avg = ts.rolling(window=7).mean() 
ax.plot(ts.index,avg,label='Rolling Average',ls = '--',lw = 2)
plt.legend()
plt.tight_layout()
plt.show()
```
